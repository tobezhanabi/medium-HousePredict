```python
def remove_USD(df):
    df['Price_USD']=df['Price_USD'].str.replace('USD', '', regex=False).str.strip().astype(float)
    return df
train = remove_USD(train_df)
test = remove_USD(test_df)
```

```python
def drop_column(df):
    df.drop(columns=['Dimensions', 'Plot Area'], inplace=True)
    return df
train = drop_column(train)
test = drop_column(test)
```

```python
test_df.head().T
```

```python
def clean_bath_balcony(df):
    df['Bathroom'] = df['Bathroom'].str.replace("> 10", '10').astype(float)
    df['Balcony'] = df['Balcony'].str.replace("> 10", '10').astype(float)
    return df
train = clean_bath_balcony(train)
test = clean_bath_balcony(test)
```

```python
def car_clean(df):
    df['Car Parking'] = df['Car Parking'].str.extract('(\d+)', expand=False).astype(float)
    return df
train = car_clean(train)
test = car_clean(test)
```

```python
test_df.head().T
```

```python
def extract_totalFloor(df):
    df['Total_Floor'] = df['Floor'].str.extract(r'(\d+)$').astype(float)
    return df
train = extract_totalFloor(train)
test = extract_totalFloor(test)
```

```python
def extract_mainFloor(df):
    df['Floor_Number'] = df['Floor'].str.extract(r'^(.+?)\s+out of').squeeze().str.replace('Upper Basement', '-1', regex=False).str.replace('Lower Basement', '-2', regex=False).str.replace('Ground', '0', regex=False).astype(float)
    return df
train = extract_mainFloor(train)
test = extract_mainFloor(test)
```

```python

train.drop(columns=['Floor'], inplace=True)
test.drop(columns=['Floor'], inplace=True)
```

```python
def clean_area(df, col):
    extracted = df[col].str.extract(r'([\d.]+)\s*(sqft|sqyrd|sqm)')
    numbers = extracted[0].astype(float)
    units = extracted[1]

    conversion = {'sqft': 1, 'sqyrd': 9, 'sqm':10.764}

    multiplier = units.map(conversion)

    df[col] = numbers * multiplier

    return df


train = clean_area(train, 'Carpet Area')
train = clean_area(train, 'Super Area')
test = clean_area(test, 'Carpet Area')
test = clean_area(test, 'Super Area')
```

### REPLACE NAN WITH MODE FOR CATEGORICAL, MEDIAN FOR NUMERIC

```python
cols_to_drop = ['Society', 'Car Parking', 'Carpet Area', 
                'overlooking', 'facing', 'Ownership', 'Description']

def drop_cols(df):
    return df.drop(columns=cols_to_drop)

train = drop_cols(train)
test = drop_cols(test)
```

######❌ Never calculate median from test and use it — that's data leakage
✅ Always learn statistics from train, apply to both

```python

numeric_cols =['Super Area', 'Balcony',  'Price_USD', 'Total_Floor', 'Floor_Number', 'Bathroom']

catrgorical_cols = ['Furnishing', 'Status', 'Transaction']

numeric_medians = train[numeric_cols].median()

categorical_modes = train[catrgorical_cols].mode().iloc[0]

def impute(df):
    df[numeric_cols] = df[numeric_cols].fillna(numeric_medians)
    df[catrgorical_cols] = df[catrgorical_cols].fillna(categorical_modes)
    return df

train = impute(train)
test = impute(test)
```
