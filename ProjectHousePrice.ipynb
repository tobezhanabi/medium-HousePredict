```python
def remove_USD(df):
    df['Price_USD']=df['Price_USD'].str.replace('USD', '', regex=False).str.strip().astype(float)
    return df
train = remove_USD(train_df)
test = remove_USD(test_df)
```

```python
def drop_column(df):
    df.drop(columns=['Dimensions', 'Plot Area'], inplace=True)
    return df
train = drop_column(train)
test = drop_column(test)
```

```python
test_df.head().T
```

```python
def clean_bath_balcony(df):
    df['Bathroom'] = df['Bathroom'].str.replace("> 10", '10').astype(float)
    df['Balcony'] = df['Balcony'].str.replace("> 10", '10').astype(float)
    return df
train = clean_bath_balcony(train)
test = clean_bath_balcony(test)
```

```python
def car_clean(df):
    df['Car Parking'] = df['Car Parking'].str.extract('(\d+)', expand=False).astype(float)
    return df
train = car_clean(train)
test = car_clean(test)
```

```python
test_df.head().T
```

```python
def extract_totalFloor(df):
    df['Total_Floor'] = df['Floor'].str.extract(r'(\d+)$')
    return df
train = extract_totalFloor(train)
test = extract_totalFloor(test)
```

```python
def extract_mainFloor(df):
    df['Floor_Number'] = df['Floor'].str.extract(r'^(.+?)\s+out of').squeeze().str.replace('Upper Basement', '-1', regex=False).str.replace('Lower Basement', '-2', regex=False).str.replace('Ground', '0', regex=False).astype(float)
    return df
train = extract_mainFloor(train)
test = extract_mainFloor(test)
```
